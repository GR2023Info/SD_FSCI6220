```python
moyenne =60000
ecart_type=12000
taille=500
donnees_de_salaires = np.random.normal(loc=moyenne, scale=ecart_type, size=taille)


#Arrondir les valeurs à 8 décimales en utilisant np.round.
donnees_de_salaires = np.round(donnees_de_salaires, decimals=0) 
donnees_de_salaires = donnees_de_salaires.tolist() # convertir le array en liste
print (donnees_de_salaires)
```

```python
sb.set_theme() 
sb.displot(donnees_de_salaires).set(title="Distribution des salaires",xlabel="salaires") 
plt.show()
```

```python
#Calculate the mean and standard deviation
mean = np.mean(donnees_de_salaires) 
print(mean)
std_dev = np.std(donnees_de_salaires)
print(std_dev)
threshold = 3

#Identify outliers 
outliers = []
for salaires in donnees_de_salaires: 
    if abs(salaires - mean)> threshold * std_dev:
       outliers.append(salaires)
       
print("Outliers:", outliers)
```
